```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
file = "bestsellers with categories.csv"
df = pd.read_csv(file)
```

```python
print("Перші п'ять рядків:")
print(df.head())
```

```python
print("\nРозміри датасету (рядків, стовпців):", df.shape)
```

//Запитання: Про скільки книг зберігає дані датасет?
//Відповідь: Датасет містить інформацію про 550 книг.

```python
df.columns = ['name', 'author', 'user_rating', 'reviews', 'price', 'year', 'genre']
```

```python
print("\nОновлені назви стовпців:")
print(df.columns)
```

```python
print("\nКількість пропусків у кожному стовпці:")
print(df.isna().sum())
```

```python
print("\nУнікальні жанри:")
print(df['genre'].unique())
```

```python
df['price'].plot(kind='hist', bins=20, edgecolor='black')
plt.title('Розподіл цін на книги')
plt.xlabel('Ціна')
plt.ylabel('Кількість книг')
plt.show()
```

```python
max_price = df['price'].max()
min_price = df['price'].min()
mean_price = df['price'].mean()
median_price = df['price'].median()
```

```python
print("\nМаксимальна ціна:", max_price)
print("Мінімальна ціна:", min_price)
print("Середня ціна:", mean_price)
print("Медіанна ціна:", median_price)
```

```python
max_rating_book = df[df['user_rating'] == df['user_rating'].max()]
max_rating_book_name = max_rating_book['name'].values[0]
```

```python
print("\nКнига з найвищим рейтингом:", max_rating_book_name)
```

```python
count_max_rating_books = len(df[df['user_rating'] == df['user_rating'].max()])
```

```python
print("Кількість книг з найвищим рейтингом:", count_max_rating_books)
```

```python
max_reviews_book = df[df['reviews'] == df['reviews'].max()]
max_reviews_book_name = max_reviews_book['name'].values[0]
```

```python
print("\nКнига з найбільшою кількістю відгуків:", max_reviews_book_name)
```

```python
max_price_2015 = df[(df['year'] == 2015) & (df['price'] < 8)].sort_values('price', ascending=True).tail(1)
max_price_2015_book_name = max_price_2015['name'].values[0]
```

```python
print("\nНайдорожча книга, яка увійшла до Топ-50 у 2015 році та коштує менше 8 доларів:", max_price_2015_book_name)
```

```python
count_fiction_2010 = len(df[(df['genre'] == 'Fiction') & (df['year'] == 2010)])
```

```python
print("\nКількість книг жанру Fiction у Топ-50 у 2010 році:", count_fiction_2010)
```

```python
count_rating_49_2010_2011 = len(df[(df['user_rating'] == 4.9) & ((df['year'] == 2010) | (df['year'] == 2011))])

print("\nКількість книг з рейтингом 4.9 у Топ-50 у 2010 та 2011 роках:", count_rating_49_2010_2011)
```

```python
sorted_books_2015 = df[(df['year'] == 2015) & (df['price'] < 8)].sort_values('price', ascending=True)
```

```python
print("\nОстання книга у відсортованому списку:", sorted_books_2015['name'].iloc[-1])
```

```python
subset_data = df[['user_rating', 'reviews']]
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=subset_data, x='user_rating', bins=10, kde=True, color='skyblue')
plt.title('Розподіл рейтингу книг')
plt.xlabel('Рейтинг')
plt.ylabel('Кількість книг')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=subset_data, x='user_rating', y='reviews', color='coral', alpha=0.7)
plt.title('Зв\'язок між рейтингом та кількістю відгуків')
plt.xlabel('Рейтинг')
plt.ylabel('Кількість відгуків')
plt.show()
```

```python
average_price_by_year = df.groupby('year')['price'].mean().reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(x='year', y='price', data=average_price_by_year, palette='viridis')
plt.title('Середня ціна книг за роками')
plt.xlabel('Рік')
plt.ylabel('Середня ціна')
plt.xticks(rotation=45)
plt.show()
```
